```python
# Ticks, Labels, Legend
# Create Object - Create Plot = Fill with randn / sum of previous element
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(randn(1000).cumsum(),'r--') # And Add a Color!!!
```

```python
# Will set the ticks here
ax.set_xticks([0,250,500,750,1000])
ax.set_xticklabels(['One','Two','Three','Four','Five'], rotation=30,fontsize='small') # Set the names of each tick
ax.set_title('My Matplotlb Plot With Ticks') # Add a title
ax.set_xlabel('Stages') # Label the Axis
ax.set_ylabel('Values')

fig
```

```python
# This is sent as a dictionary
props = {'title': 'My First Matplotlip Plot with Ticks',
        'xlabel': 'Stage',
        'ylabel': 'Values'}
ax.set(**props)
fig
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)

ax.plot(randn(1000).cumsum(), 'k-',label='one') # Random Numbers (1000 of them), & cumulative sum of data & color, Label
ax.plot(randn(1000).cumsum(),'g--',label='two')
ax.plot(randn(1000).cumsum(),'y.',label='three') # Line style is dots
#Creating a legend and description

ax.legend(loc='best') # Place it at the best location
```
