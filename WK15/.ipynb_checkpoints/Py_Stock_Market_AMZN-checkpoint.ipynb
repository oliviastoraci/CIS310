```python
# COunt the number of numerical and non-numerical variables (columns)
# N tells us number of columns
from pandas.api.types import is_numeric_dtype

n=0

for c in data.columns:
    if is_numeric_dtype(data[c]):
        n+=1

print('Numerical Variables: ', n, 'Non-Numerical Variables: ', len(data.columns)-n)
```

```python
# Data Cleaning - Preparing Data for Analytics
data.isnull()
```

```python
# Identify the sum of the null values in the data set. Num missing vals in each row
data.isnull().sum()
```

```python
# Handle missing Data
# See the data being dropped - 
data.dropna()
```

```python
# Best way to drop data and change the data. Change the data itself
data.dropna(inplace=True)
```

```python
# Now see the permanent data without any null values
data.isnull().sum()
```
